<h2 align='center'>Abstract</h2>

The "Black Death" is an important historical event, that took place in the XIVth century in Europe. It is an <i>outbreak</i> of a disease called the <i>bubonic plague</i> that impact Europe and resulted in the death of roughly half of the European population over the course of five years.

<h2 align='center'>The bubonic plague: an overview</h2>

The bubonic plague is a very serious <i>infectious disease</i> that has been impacting human population for above two thousand years now. Infectious means that it is caused by a <i>pathogen</i>, i.e. a micro-organism that enters the human body and develops within the body, disrupting its normal way to function. This result in an <i>infection</i> which, in the case of the bubonic plague, is likely to be fatal unless it is treated by appropriate medications, a plague vaccine. 


The pathogen responsible of the bubonic plague is a <i>bacteria</i> called <i>Yersinia pestis</i>. This name is composed of the latin name of the plague, <i>pestis</i> prefixed by a latin version of the name of the person who first discovered the bacteria, Alexander Yersin.

The infection caused by <i>Yersinia pestis</i> causes flu-like symptoms such as headaches and a high fever, a common symptom of many infections, together with visible sympetoms such as the apparition in some specific parts of the body of swelling (named <i>bubos</i>, which gave the name bubonic to the plague) and the development of black spots on the skin, which motivated the "Black Death" name. In general, an untreated patient dies within a few days. 

<h2 align='center'>Infectious diseases</h2>

Infectious diseases, as mentionned above, are caused by pathogens. There are various types of pathogens, including viruses, bacterias and parasites. Viruses causes for example diseases such as the flu (<i>influenza</i>), measles or mumps. Bacterias are reponsible for many diseases such as pneumonia, tuberculosis, or plague. Last parasites are larger micro-organisms, such as roundworms for example; examples of parasitic diseases include malaria or the dengue.

<h2 align='center'>How does the bubonic plague pathogen infect humans?</h2>

An individual becomes sick with the buonic plague when the pathogen agent <i>Yersinia pestis</i> enters its body. The most common way this happens is when the individual is bitten by a <i>flea</i> that carries tha pathogen in its gut and injects it when biting (it is a little more complicated actually). The flea is called a <i>vector</i> of the disease: it does not cause the disease by itsdelf, but helps the pathogen to move from the environment to within an individual. 

It is widely accepted that the kinds of fleas that carry <i>Yersinia pestis</i> are not human fleas (i.e. the kind of fleas that naturally feed on human blood) but <i>rodent</i> fleas, that usually feed on the blood of rats, mice, squirrels, prairie dogs and so on. So why do they suddenly bite humans ? This is because the rodents they use to feed on are also infected by the plague agent, and so they die, forcing the fleas to look for other sources of food. If the concerned rodent population (called the <i>reservoir</i> of the disease) lives close to human populations (like urban rats), then the natural new hosts are humans. <h2 align='center'>Infectious diseases outbreaks</h2>

An outbreak is when a disease, that might have been present or not within a population, suddenly starts to infect more people than usual in a given place (in general city, region or country). When an outbreak spreads over a larger region it is called an <i>epidemic</i>, or even a <i>pandemic</i> when it happens over a very large region.

Outbreaks occur very often. We can think for example to the annual flu outbreak that hits every canadian city every winter. This is not too scarry and the local health authorities know how to handle it, through a campaign of flu vaccination. But even now more serious outbreaks occur quite often, such as the Ebola epidemic in Western Africa in 2013-2016 that caused thousands of death and was very challenging for local and global health authorities.

There has been many other dramatic outbreaks, epidemic and pandemics in history, the Black Death that we will describe below being one of the most known. But closer to us, we can think about several smallpox epidemic in the late XiXth century that spread through first nations communities of the Pacific NorthWest. 

<h2 align='center'>The Black Death Timeline</h2>

The Black Death was a bubonic plague pandemic, that concerned all of Western Europe. It started outide of Europe, in Asia, where it infected soldiers of a Mongol army who were besieging a Genoese trapest in Crimea, Kaffa. After the siege, the Genoese traders (from the Italian city of Geona) sailed back home without knowing they were themselves infected. This is how the disease was introduced in Western Europe, through Italy. 

At the time people did not know the principles of infectious diseases and did not intervene quickly to try to conceal the disease to a city or a few cities where the outbreak(s) would have ended likely quickly. Instead, unaware of the dramatic situation that was unfolding, traders from Italian cities maintained very active relationships with the rest of Europe, through marine or land trading routes. This resulted in the disease spreading very fast to Southern Europe within 6 months, and from there turning into a European-wide pandemic within three years.

<h2 align='center'>Historical plague outbreaks</h2>

The Black Death was not the only major plague pandemic. Before this medieval pandemic, in the Vth century (in 541-542 precisely), the <i>Justinian Plague</i> developed as an epidemic in the  Eastern part of the Mediterannean region and many Mediterranean port cities. It is known as the first major plague pandemic.

The Black Death itself lasted much longer than the few years illustrated in the map above. The years 1347-1351 correspond to the most acute part of the pandemic. But after this, plague remained present in Europe for almost 400 years, flaring up here in there in dramatic outbreaks, although none that turned into a pandemic. People had learned from the Black Death, and as soon as plague reappeared in a community, the community was isolated to prevent the disease spread. nevertheless, some important outbreaks impacted large cities such as the Great Plague of London (1665-1666) or the Marseille Plague (1722). This very long pandemic is known as the second major plague pandemic.

Past the Marseille plague, the disease seem to disappear from Europe, although it remained present in Asia. This is there, in the Chinese province of the Yunan, in 1855 precisely, that the third major plague pandemic started. It did not move out of China for a long time, but then reached Hong-Kong in 1894, and from there, following again the maritime trade routes, it reached the rest of the world and was considered as an active pandemic by the World Health Organisation until 1959.

Currently, there is no bubonic plague pandemic, but outbreaks occur regularly. The most serious ones are in the African island of Madagascar, but there are also regular small outbreaks in the South-West of the United States, where prairie dogs act as a reservoir for the disease.

<h2 align='center'>Was the Black Death a bubonic plague outbreak</h2>

This is actually a good question. Indeed, nobody did the same work than Yersin and  Shibasaburō during the Black Death pandemic. Bacterias were actyually not known at the time. So in order to declare it a plague, historians relied on written records from people chronicling the dramatic pandemic. Nevertheless, most chroniclers were not physicians and as such they did not use the medical vocabulary developed in the XIXth century when the mechanims of infectious diseases were discovered. It did not help that, at the time of the Black Death, the word "plague" was used as a generic word used to describe infectious diseases outbreaks in general.

As a result of this lack of precise documentation about the Black Death pandemic, there was a long-standing debate abut the cause of this pandemic, with a strongly supported hypothesis that the disease was actually due to an Ebola-like virus. 

It is only in 2011, that an international team led by a Canadian scientist, Hendrik Poinar, from McMaster University, was able to extract, from the dental pulp of the body of victims of the Black Death in London, the genome of the bacteri <i>Yersinia pestis</i>, thus closing the debate over the actual disease that caused the Black Death.

<h2 align='center'>Modelling a Bubonic Plague Outbreak</h2>

<h4 align = 'center'> $\mid$ Grade 11 $\mid$ Social Sciences $\mid$</h4>

We are interested in modelling a bubonic plague outbreak. We part from the assumption that the total polulation can be subdivided into a set of classes, each of which depends on the state of the infection. The **SIR Model** is the simplest one and, as its name suggests, it divides the population into three classes. 

<h2 align='center'>The SIR Model: Principle</h2>

### Population Parameters

In this model, the total population is divided into three groups of individuals:

* Susceptible: the individuals that are not infected but can become infected
* Infected: the individuals that are infected
* Removed: the individuals that are either dead (the pessimistic point of view) or survived 

We are concerned about parameters that track the **number of people that are susceptible**, the **number of people infected**, and the **number of people who are removed**. We denote these parameters to be $S, I, R$ respectively. In other words we want to understand how, when time passes, the number of individuals in each group evolves. If we have a realistic model, then it might be useful to predict the long-term outcome of an outbreak and inform public health interventions.

For example if we can predict that the number fo removed people will stay low and the number of infected people will quickly go down to zero, then there is no need to intervene and we can let the outbreak ends by itself. Conversely if we see a large and quick increase of the numbers of infected and removed individuals, then the outbreak needs a quick intervention to be tamed.

We now describe the SIR model as a mathematical model.



We write $S_n, I_n, R_n$ to denote the number, respectively, of susceptible, infected and removed individuals at time point $n$

In this model, we assume that the total population is constant (so births and deaths are ignored) for the duration of the model simulation. We denote the **total population** by $N$, and write  $$N=S_n + I_n + R_n.$$

### Modelling theoutbreak progression

We assume that an susceptible individual can become infected only by contact with an already infected individual. Once infected, we assume that the disease takes a constant amount of time to progress within an individual until he/she is removed (i.e. either dies or recovers). So there are two processes in play: infection and removal. We need to define these processes and model how they impact the transition from a state $(S_n,I_n,R_n)$ to the next state $(R_{n+1},I_{n+1},R_{n+1})$.

<img src="./images/SIR_1.jpg" width="750px"/>

The disease parameters we are concerned about are:
* Rate of infection, equivalently the rate of contact of a susceptible individual with an infected individual

* Rate of removal of an infected individual.

The **rate of infection**, denoted by $\beta$, is defined to be the probability of getting the disease when a contact is made between a susceptible and infectious subject. In the diagram below, the transition rate between the compartment of the susceptible population and the infected population is given by $\beta \cdot I$


The **rate of removal**, denoted by $\\gamma$, is defined to be rate of removal (recovery or death). If we are given that the duration of an infection is $T$, then $\gamma = \dfrac{1}{T}$. We can understand it as the proportion of currently infected individuals that will be removed at the next time point.

<img src="./images/SIR_2.jpg" width=\"500px\">


Taking into account the rate of contact $\\beta$ and rate of removal $\gamma$, then each group population changes within one unit of time as follows


$$
\begin{align}
S_{n+1} &= S_n - {\beta} S_n I_n\\
I_{n+1} &= I_n + {\beta} S_n I_n - \gamma I_n \\
R_{n+1} &= R_n + \gamma I_n\\
N&=S_n + I_n + R_n
\end{align}
$$

These equations form the SIR model. They allow, from knowing the parameters of the model (i.e. the characteristics of the disease at play) and the current state $(S_n,I_n,R_n)$ of a population to predict the next states of the population for later time points. Such models are critical in our days for monitoring and controlling infectious diseases outbreaks.

### Parameters Summary

$$\begin{align}
\text{S} &= \text{Number of people susceptible} \\
\text{I} &= \text{Number of people infected} \\
\text{R} &= \text{Number of people removed} \\
\beta &= \text{Infection Rate} \\
\gamma &= \text{Removal Rate} \\
\text{N} &= \text{Total population (we assume N is constant)} 
\end{align}$$

Use the widgets below to simulate the plague outbreak of Eyam in 1665.

```python
# This function takes as input a vector y holding all initial values,
#    n the number of weeks the infection runs for, N total population
#    beta rate of contact, gamma rate of removal

def discrete_SIR(S,I,R,n,N, beta, gamma):
     
    for i in range(n):
        S_next = S[i] - beta*((S[i]*I[i])/N)
        S.append(S_next)
        
        I_next = I[i] + beta*((S[i]*I[i])/N) - gamma*I[i]
        I.append(I_next)
        
        R_next = R[i] + gamma* I[i]
        R.append(R_next)
    return [S,I,R]
import matplotlib.pyplot as plt
import numpy as np
from math import ceil

def plot_SIR(S_i,I_i,R_i, n,beta,gamma,original_data):
    
    S,I,R = [S_i],[I_i],[R_i]
    
    fig = plt.figure(facecolor='w',figsize=(17,5))
    ax = fig.add_subplot(111,facecolor = '#ffffff')
    N = S_i + I_i + R_i
    
    data = discrete_SIR(S,I,R,n,N,beta,gamma)
    S_f = data[0]
    I_f = data[1]
    R_f = data[2]
    #fig = plt.figure(facecolor='w',figsize=(10,5))
    #ax = fig.add_subplot(111,facecolor = '#dddddd')
    if original_data==False:
    
        x = [i for i in range(n+1)]

        
        plt.scatter(x,S_f,c= 'b',label='Susceptible')
        plt.scatter(x,I_f,c='r',label='Infected')
        plt.scatter(x,R_f,c='g',label='Removed')
    else:
        x = [i for i in range(n+1)]

        
        plt.scatter(x,S_f,c= 'b',label='Susceptible')
        plt.scatter(x,I_f,c='r',label='Infected')
        plt.scatter(x,R_f,c='g',label='Removed')
        original_i = [7,14,22,29,21,8,21,0]
        original_s = [254,235,201,153,121,108,121,83]
        ori_days = [0,14,28,42,56,70,84,112]

        x_ran = [i for i in range(ori_days[-1] + 1)]
        points = [(ori_days[i],original_i[i]) for i in range(len(ori_days))]

        plt.scatter(ori_days,original_i,c='black',label="Original data:Infected")
        plt.scatter(ori_days,original_s,c='orange',label="Original data:Susceptible")
    
    plt.xlabel('Time (days)')
    plt.ylabel('Number of individuals')
    plt.title('Simulation of the plague outbreak in Eyam, 1665')
    legend = ax.legend()
    plt.show()
    print("SIMULATION DATA\n")
    print("Rate of infection: " + str(beta))
    print("Rate of removal: " + str(gamma))
    print("\n")
    print("Initial Conditions:")
    print("Total number of Susceptible: "  + str(ceil(S_f[0])))
    print("Total number of Infected: "  + str(ceil(I_f[0])))
    print("Total number of Removed: "  + str(ceil(R_f[0])))
    print("\n")
    print("After " + str(n) + " days:")
    print("Total number of Susceptible: "  + str(ceil(S_f[-1])))
    print("Total number of Infected: "  + str(ceil(I_f[-1])) )
    print("Total number of Removed: "  + str(ceil(R_f[-1])))
# Tweaking initial Values
from ipywidgets import widgets, interact, interact_manual

s = {'description_width': 'initial'}
interact_manual(plot_SIR,
        S_i  =widgets.IntSlider(value = 254,
                               min = 200,
                               max = 1000,
                               step = 1,
                               style=s,
                               description="Susceptible Initial",
                               disabled=False,
                               orientation='horizontal',
                               readout=True),
        I_i = widgets.IntSlider(value = 7,
                               min = 0,
                               max = 500,
                               step = 1,
                               style=s,
                               description="Infected Initial",
                               disabled=False,
                               orientation='horizontal',
                               readout=True),
         R_i = widgets.IntSlider(value = 0,
                               min = 0,
                               max = 500,
                               step = 1,
                               style=s,
                               description="Removed Initial",
                               disabled=False,
                               orientation='horizontal',
                               readout=True),
         n = widgets.IntSlider(value = 112,
                               min = 0,
                               max = 500,
                               step = 1,
                               style=s,
                               description="Time (days)",
                               disabled=False,
                               orientation='horizontal',
                               readout=True),
         beta = widgets.FloatSlider(value = 0.14909440503418078,
                            min = 0,
                            max = 1.0,
                            step = 0.1,
                            style = s,
                            description = 'Infection Rate',
                            disabled=False,
                            orientation='horizontal',
                            readout=True),
        gamma = widgets.FloatSlider(value = 1/11,
                             min = 0,
                            max = 1.0,
                            step = 0.1,
                            style = s,
                            description = 'Removal Rate',
                            disabled=False,
                            orientation='horizontal',
                            readout=True),
        original_data = widgets.Checkbox(
    value=False,
    description='Show Original Data',
    disabled=False
)
        );
```

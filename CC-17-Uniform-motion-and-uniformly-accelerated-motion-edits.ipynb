##  Uniformly Accelerated Motion

<s> The motion should be uniformly accelerated when the rate of increase of velocity of a moving object in a particular direction
is maintained constant all the time. Let's think, you are running to catch a ball. Your motion of running is uniformly accelerated. You able to catch the ball after $5 \  \textrm{sec}$ and your acceleration is $2 \ \textrm{ms}^{-2}$. As you are uniformly accelerated, your acceleration will be constant throught out this $5 \  \textrm{sec}$. That means your velocity will be increased by $2 \ \textrm{ms}^{-1}$ at each second. Let's see this uniformly accelerated motion scenario in below plot for this example.</s>

<font color='blue'> Uniformly accelerated motion is a little different than the uniform motion we discussed earlier. In the case of uniformly accelerated motion, your velocity is increasing constantly and equally in time, in a way analogous to how displacement changes if you're traveling at constant velocity. For example, suppose you start at rest and begin running in order to catch a ball. In this scenario, you will need to _accelerate_. Suppose after each second, you are traveling 2 $\frac{m}{s}$ faster than the second previous. In such a case, you have a uniform acceleration of 2 $\frac{m}{s^2}$ (the seconds are squared in the units of acceleration as you are increasing your velocity constantly - i.e. you gain more velocity per second). Uniform acceleration is demonstrated in the graphs below. </font>

\begin{equation}
a=\frac{v-u}{t} \\

\Rightarrow v=u+at

### $v = u^{2}+2as$
<s>This is a rearranging motion equation. </s> <font color='blue'> This equation is actually not anything new - it is just the combination of two of the equations we've already seen! </font> We can get this equation by rearranging the the equations $v=u+at$ and $s = ut+\frac{1}{2}at^{2}$, <font color='blue'> as follows </font>

\begin{equation}
v=u+at \\
\Rightarrow v-u=at \\

\begin{equation}

### Problem 4

An Air Canada plane <font color = 'blue'> requires </font> <s> has</s> a takeoff speed of $78.5 \ ms^{-1}$ and <s>requires</s> $1690 \ m$ <s>running</s> <font color = 'blue'> of runway </font> to reach that speed. Determine the acceleration of this plane and the time required to reach this speed.

#### Solution

**Step 1:**

\begin{equation}
\textrm{Given}, \\
\textrm{initial} \ \textrm{velocity}, u = 0 \ ms^{-1} \\
\textrm{final} \ \textrm{velocity}, v = 78.5 \ ms^{-1} \\
\textrm{displacement}, s = 1690 \ m \\
\textrm{acceleration}, a=? \\
\textrm{time},t=?
\end{equation}

**Step 2:**

\begin{align}
\textrm{Motion} \ \textrm{equation},  v^{2}=u^{2}+2as \\
v=u+at
\end{align}

**Step 3:**

Determine acceleration,

\begin{equation}
 v^{2}=u^{2}+2as \\
\Rightarrow a=\frac{v^{2}-u^{2}}{2a} \\
\Rightarrow a=\frac{78.5^{2}-0^{2}}{2\times1690} \\
 \Rightarrow a=1.82 \ ms^{-2}
\end{equation}

Determine time, 

 \begin{equation}
 v=u+at \\
 \Rightarrow t=\frac{v-u}{a} \\
 \Rightarrow t=\frac{78.5-0}{1.82} \\
 \Rightarrow t = 43.06 \ s
 \end{equation}
 
 **Step 4:**

\begin{align}
\textrm{acceleration},a=1.82 \ ms^{-2}\ (Ans) \\
\textrm{time}, t = 43.06 \ s \ (Ans)
\end{align}

```python
import folium
from folium.plugins import MarkerCluster
north_map = folium.Map(location=[53.5,-114],zoom_start=5)
mc = MarkerCluster()
for row in df[df['Latitude']>53.225].itertuples():
  mc.add_child(folium.Marker(location=[row.Latitude,row.Longitude], tooltip=row.Well_ID))
north_map.add_child(mc)
north_map
```

```python
import folium
from folium.plugins import MarkerCluster
lat = 53
lon = -113.5
ne_map = folium.Map(location=[lat,lon],zoom_start=5)
mc = MarkerCluster()
for row in df[(df['Latitude']>lat)&(df['Longitude']>lon)].itertuples():
  mc.add_child(folium.Marker(location=[row.Latitude,row.Longitude], tooltip=row.Well_ID))
ne_map.add_child(mc)
ne_map
```

# Creating, Labelling, and Interpreting Line Graphs
Line graphs are useful for visualizing (seeing) patterns in data. [Wikipedia](https://en.wikipedia.org/wiki/Line_chart) defines them as "a type of chart which displays information as a series of data points called 'markers' connected by straight line segments."

We can make line graphs using Python. We need to import the "matplotlib" library, say what data to graph, and maybe add some labels.

The horizontal axis is the x-axis, and the vertical axis is the y-axis. Lets put the x values in a list called **x** and the y values in a list called **y**. Select the cell below this one and click the `Run` button on the toolbar to display a graph.

```python
import matplotlib.pyplot as plt
%matplotlib inline
x = [1, 2, 3]
y = [5, 6, 7]
plt.plot(x, y)
plt.show()
```

Now let's try some real-world data, the [population of Canada over time](https://en.wikipedia.org/wiki/Population_of_Canada#Canada_as_a_whole_since_confederation). Population will go on the y-axis and time on the x-axis. `Run` the next cell.

```python
population = [3463000, 4883000, 8060000, 11654000, 20500000, 28377000]
year = [1867, 1892, 1917, 1942, 1967, 1992]
plt.plot(year, population)
plt.show()
```

We can see that the line gets steeper, meaning that the rate of population growth was increasing.

Now let's add some labels to that graph so other people will know what we are displaying.

```python
populations = [3463000, 4883000, 8060000, 11654000, 20500000, 28377000]
years = [1867, 1892, 1917, 1942, 1967, 1992]
plt.plot(years, populations)
plt.title('Population of Canada Over Time')
plt.ylabel('population')
plt.xlabel('year')
plt.show()
```

Here is a graph of average daytime high temperature in Edmonton, Alberta by month (using data from [NOAA](https://www.ncdc.noaa.gov/)). 

```python
y = [-7, -3, 2, 11, 18, 21, 23, 22, 17, 11, 1, -5]
x = list(range(1,13))

plt.plot(x, y)
plt.title('Average Temperature in Edmonton by Month')
plt.ylabel('Temperature (°C)')
plt.xlabel('Month')
plt.xlim(1,12)
plt.grid(True)
plt.show()
```

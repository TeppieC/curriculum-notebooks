# Introduction

Everything in the universe is moving. It can be very very slow or small amount, but the movement is happening. Suppose, you are just standing, but you are still moving as the earth is moving around the sun and the sun is moving around galaxy. Motion never stops. So, the motion is an important topic of physics. Motion defines the change of position of an object with respect to other surrounding objects e.g. a car is moving with respect to trees on the roadside. The motion relates to the velocity, speed and acceleration. In this notebook, we will familiar with two types of motion. One is uniform motion and another one is uniformly accelerated motion. 

# Background

## Fundamental Concept

As we mentioned in the introduction part that motion is related to distance, time, velocity and acceleration. In this section, we will discuss these fundamental concepts of the motion shortly. If you are familiar with these concepts you can ignore this part.

**Distance and Displacement:** Distance defines the change of position of an object. Let's say, you ran $50 \ \textrm{m}$ from your school gate. Then your distance will be $50 \ \textrm{m}$ from the school gate. Now if you say that I ran $100 \ \textrm{m}$ north from the school gate. Then this $100 \ \textrm{m}$ will be your displacement. So, displacement defines the change of position of an object in a definite direction.

**Speed:** Speed defines the rate of change of position of an object. In other way, we can say that the quantity by which we can measure how fast an object moves or distance traveled, is called speed. For the above example, speed will be:

\begin{equation}
\textrm{speed} = \frac{\textrm{change} \ \textrm{of} \ \textrm{distance}}{\textrm{time}} = \frac{100}{20} = 5\ ms^{-1}
\end{equation}

**Velocity:** Velocity defines the rate of change of position in a definite direction. In another way, we can say that rate of change of displacement is known as velocity. From the above example, we have found that your displacement is $100 \ \textrm{m}$. If you take $20 \ \textrm{sec}$, then your velocity will be: 

\begin{equation}
\textrm{velocity} = \frac{\textrm{change} \ \textrm{of} \ \textrm{displacement}}{\textrm{time}} = \frac{100}{20} = 5 \ ms^{-1}
\end{equation}

**Acceleration:**: Acceleration defines the rate of change of velocity of an object. Then the acceleration of the above example will be:

\begin{equation}
\textrm{acceleration} = \frac{\textrm{change} \ \textrm{of} \ \textrm{velocity}}{\textrm{time}} = \frac{5}{20} = 0.25 \  ms^{-2}
\end{equation}

Uniform motion describes the motion when an object travels in a straight line, and the object travels equal distance in equal interval of time. In a simple way, we can say that when an object travels a fixed or constant distance in each time interval then the motion will be uniform. Let's think every day you go to your school by walk. The distance of school is $80 \ \textrm{m}$ straight from your home. You need to walk $40 \ \textrm{sec}$ to reach the school. You travel $20 \ \textrm{m}$ in first $10 \ \textrm{sec}$. Then, motion will be uniform when you travel $20 \ \textrm{m}$ in each $10 \ \textrm{sec}$ time interval. Let's see the plot of velocity and displacement in this uniform motion scenario:

```python
%%capture
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.animation
plt.rcParams["animation.html"] = "jshtml"
import numpy as np

# create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(6, 3.25), dpi= 200, facecolor='w', edgecolor='k')

# Same X axis limt and grid initalizations
for ax in [ax1, ax2]:
    ax.set_xlim(0, 40)
    ax.grid()
    
ax1.set_ylim(0,80)
ax2.set_ylim(0,5)    

# Data
t = np.linspace(0,40,5)
d = np.linspace(0,80,5)
v = np.linspace(2,2,5)

# initialize the plot
l1, = ax1.plot([],[], 'go-', label='Displacement', linewidth=2)
leg = ax1.legend(loc='best')
l2, = ax2.plot([],[],  'rs-', label='Velocity')
leg = ax2.legend(loc='best')

fig.suptitle('Uniform Motion')
ax1.set_xlabel('Time (sec)')
ax2.set_xlabel('Time (sec)')

# Animate the plot
def animate(i):
    l1.set_data(t[:i+1], d[:i+1])
    l2.set_data(t[:i+1], v[:i+1])
    
ani = matplotlib.animation.FuncAnimation(fig, animate, frames=len(t))
```

##  Uniformly Accelerated Motion

The motion should be uniformly accelerated when the rate of increase of velocity of a moving object in a particular direction
is maintained constant all the time. Let's think, you are running to catch a ball. Your motion of running is uniformly accelerated. You able to catch the ball after $5 \  \textrm{sec}$ and your acceleration is $2 \ \textrm{ms}^{-2}$. As you are uniformly accelerated, your acceleration will be constant throught out this $5 \  \textrm{sec}$. That means your velocity will be increased by $2 \ \textrm{ms}^{-1}$ at each second. Let's see this uniformly accelerated motion scenario in below plot for this example.  

```python
%%capture
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.animation
plt.rcParams["animation.html"] = "jshtml"
import numpy as np

# create a figure with two subplots
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(10, 3.25), dpi= 600, facecolor='w', edgecolor='k')

# Same X axis limt and grid initalizations
for ax in [ax1, ax2, ax3]:
    ax.set_xlim(0, 5)
    ax.grid()
    
ax1.set_ylim(0,25)
ax2.set_ylim(0,10)    
ax3.set_ylim(0,5)

# Data
t = np.linspace(0,5,6)
d = ([0,1,4,9,16,25])
v = ([0,2,4,6,8,10])
a = np.linspace(2,2,6)

# initialize the plot
l, = ax1.plot([],[], 'go-', label='Displacement', linewidth=2)
leg = ax1.legend(loc='best')
l1, = ax2.plot([],[],  'rs-', label='Velocity')
leg = ax2.legend(loc='best')
l2, = ax3.plot([],[], 'b*-', label='Acceleration', linewidth=2)
leg = ax3.legend(loc='best')

fig.suptitle('Uniformly Accelerated Motion')
ax1.set_xlabel('Time (sec)')
ax2.set_xlabel('Time (sec)')
ax3.set_xlabel('Time (sec)')

# Animate the plot
def animate(i):
    l.set_data(t[:i+1], d[:i+1])
    l1.set_data(t[:i+1], v[:i+1])
    l2.set_data(t[:i+1], a[:i+1])
    
ani = matplotlib.animation.FuncAnimation(fig, animate, frames=len(t))
```

## Equations of Motion

We can solve the motion related moving object problem using four equations. These equations are known as equations of motion. These equations are applicable when the object moves along a straight line with uniform acceleration. Let an object with initial velocity $\textrm{u}$ moves with uniform acceleration $\textrm{a}$. After travelling a distance $\textrm{s}$ in time $\textrm{t}$ its final velocity becomes $\textrm{v}$.  Then, we can express the equation of motion in following way:

\begin{align}
    v = u+at \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  \ \ \ \ \ \ 
    s = (\frac{u+v}{2})t
\end{align}



\begin{align}
s = ut+\frac{1}{2}at^{2} \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  \ \ 
v = u^{2}+2as
\end{align}

### $v=u+at$
This equation defines the relation between velocity and acceleration. Let's an object is moving in a constant acceleration $\textrm{a}$. From the definition of acceleration, we can write that -

\begin{equation}
a=\frac{\Delta v}{\Delta t}
\end{equation}

where $\Delta v$ is change of velocity over $\Delta t$ time period.

If the object starts moving at $t=0$ and and initial velocity is $u$. After time $t$,the velocity becomes $v$. The change of velocity is $\Delta v = v - u$ and change of time is $\Delta t = t$. Then acceleration will be - 

\begin{equation}
a=\frac{v-u}{t} \\
\Rightarrow v-u=at \\
\Rightarrow v=u+at
\end{equation}

### $s = (\frac{u+v}{2})t$
Average speed of an object, 
\begin{equation}
\textrm{average} \ \textrm{speed} = \frac{\textrm{distance} \ \textrm{covered}}{\textrm{time}} \\
\Rightarrow \frac{u+v}{2}=\frac{s}{t} \\
\Rightarrow s = (\frac{u+v}{2})t 
\end{equation}

### $s = ut+\frac{1}{2}at^{2}$

Let's an object is moving in a uniform acceleration $\textrm{a}$ with initial velocity $u$. After $t \ sec$. After $t \ sec$, the object covers distance $s$ at final velocity $v \ ms^{-1}$. The time-velocity graph in uniform acceleration scenario will be like this:

<img src="motion-graph.jpg" alt="Callysto" width=500 align = "left">

In the figure, we have split the shaded area into a triangle and a rectangle. The sum of area 1 and area 2 is the displacement. So, we have to calculate the area of triangle and rectangle. Let's calculate the area individually,

**Area 1**
 
\begin{equation}
\textrm{Area}= \textrm{length} \times \textrm{height}= \textrm{ut}
\end{equation}

**Area 2** 
\begin{equation}
\textrm{Area}= \frac{1}{2} \times {\textrm{base}} \times {\textrm{height}} \\
\Rightarrow \textrm{Area}= \frac{1}{2} \times {t} \times {(v-u)} \\
\end{equation}

As we know, $(v-u) = at$. Replacing $(v-u)$ in above equation:

\begin{equation}
\Rightarrow \textrm{Area}= \frac{1}{2} \times {t} \times {at} \\
\Rightarrow \textrm{Area} = \frac{1}{2} \times {at^{2}}
\end{equation}

The total displacement $s$ will be sum of area 1 and area 2.

\begin{equation}
s=\textrm{area 1} + \textrm{area 2} \\
\Rightarrow s = ut + \frac{1}{2} {at^{2}}
\end{equation}

### $v = u^{2}+2as$
This is a rearranging motion equation. We can get this equation by rearranging the the equations $v=u+at$ and $s = ut+\frac{1}{2}at^{2}$.

\begin{equation}
v=u+at \\
\Rightarrow v-u=at \\
\Rightarrow t=\frac{v-u}{a}
\end{equation}

Now, inserting the value of $t$ in displacement motion equation:

\begin{equation}
s = ut+\frac{1}{2}at^{2} \\
\Rightarrow s= u\frac{v-u}{a}+ \frac{1}{2} (\frac{v-u}{t})^{2} \\
\Rightarrow s= \frac{1}{2} \frac{2uv-2u^{2}+(v-u)^{2}}{a} \\
\Rightarrow s= \frac{1}{2} \frac{2uv-2u^{2}+v^{2}-2uv+u^{2}}{a} \\
\Rightarrow s=\frac{1}{2} \frac{v^{2}-u^{2}}{a} \\
\Rightarrow v^{2}-u^{2}=2as \\
\Rightarrow v^{2}=u^{2}+2as
\end{equation}

### Problem 1

A bus starting from resting state moves with acceleration of $4  ms^{-2}$ and its velocity reaches at $40 \ ms^{-1}$ . How long time does the bus take part in the acceleration activity?

#### Solution

**Step 1: Identify what information is given and what is asked for?**:

Given,

initial velocity, $u = 0 \ ms^{-1}$ 

final velocity, $v = 40 \ ms^{-1}$ 

acceleration, $a = 4 \ ms^{-2}$

time, $t = ?$

**Step 2: Find an equation of motion relating the given information to the acceleration**

In this problem, the value of $u$, $v$ and $a$ are given and $t$ is required. Now, check the above $4$ motion equations, find the equation is related to $u$, $v$, $a$ and $t$. We find the equation and it is $v=u+at$.

**Step 3: Ensure that all the values are in the correct units and fill them in the selected equation.**
After checking all the units of the given variable $u$, $v$ and $a$, we find that they are i correct unit. Then fill them in selected equation:

\begin{equation}
v=u+at \\
\Rightarrow at = v -u \\
\Rightarrow t = \frac{v-u}{a} \\
\Rightarrow t = \frac{40ms^{-1}-0ms^{-1}}{4ms^{-2}} \\
\Rightarrow t = 10s
\end{equation}

**Step 4: Quote the answer and check the units.**

time = $10s$ $(Ans)$

### Problem 2

A car accelerates uniformly from $16 \ ms^{-1}$ to $38.8 \ ms^{-1}$ in $3.1 \ seconds$. Calculate the distance traveled by the car.

#### Solution

**Step 1:**

\begin{equation}
\textrm{Given}, \\
\textrm{initial} \ \textrm{velocity}, u = 16 \ ms^{-1} \\
\textrm{final} \ \textrm{velocity}, v = 38.8 \ ms^{-1} \\

s = (\frac{u+v}{2})t \\

\end{equation}

\begin{align}

### Problem 4

A Air Canada plane has a takeoff speed of $78.5 \ ms^{-1}$ and requires $1690 \ m$ running to reach that speed. Determine the acceleration of this plane and the time required to reach this speed.

#### Solution

**Step 1:**

\begin{equation}
\textrm{Given}, \\
\textrm{initial} \ \textrm{velocity}, u = 0 \ ms^{-1} \\
\textrm{final} \ \textrm{velocity}, v = 78.5 \ ms^{-1} \\
\textrm{displacement}, s = 1690 \ m \\
\textrm{acceleration}, a=? \\
\textrm{time},t=?
\end{equation}

**Step 2:**

\begin{align}
\textrm{Motion} \ \textrm{equation},  v^{2}=u^{2}+2as \\
v=u+at
\end{align}

**Step 3:**

Determine acceleration,

\begin{equation}
 v^{2}=u^{2}+2as \\
\Rightarrow a=\frac{v^{2}-u^{2}}{2a} \\
\Rightarrow a=\frac{78.5^{2}-0^{2}}{2\times1690} \\
 \Rightarrow a=1.82 \ ms^{-2}
\end{equation}

Determine time, 

 \begin{equation}
 v=u+at \\
 \Rightarrow t=\frac{v-u}{a} \\
 \Rightarrow t=\frac{78.5-0}{1.82} \\
 \Rightarrow t = 43.06 \ s
 \end{equation}
 
 **Step 4:**

\begin{align}
\textrm{acceleration},a=1.82 \ ms^{-2}\ (Ans) \\
\textrm{time}, t = 43.06 \ s \ (Ans)
\end{align}

## Equations of Motion

<del> Running, driving a car and even simply taking a walk are all our daily life motion examples. </del> 

Using our relationships between displacement, velocity, acceleration and time, we can define "equations of motion" for a moving object. There are four such equations relevant to the principles of uniform motion and uniform acceleration.

Suppose an object with initial velocity $\vec{\textrm{v}}_i$ $\textrm{ms}^{-1}$ is travelling with uniform acceleration $\vec{\textrm{a}}$ $\textrm{ms}^{-2}$. After traveling a distance $\textrm{s}$ in time $\textrm{t}$ the object's final velocity would be $\vec{\textrm{v}_f}$ $\textrm{ms}^{-1}$, these quantities are described by the following equations


<font color = 'red'> **Here you've defined both displacement and speed as $s$. This is incredibly confusing with your notation later. You will need to change the displacement to a different variable, and include a vector over it. This change will need to be made in _all_ the equations below where applicable ** </font> 

\begin{align}
   \vec{\textrm{v}_f}=\vec{\textrm{v}_i}+\vec{\textrm{a}} \textrm{t} \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  \ \ \ \ \ \ 
   \textrm{s} = (\frac{\vec{\textrm{v}_i}+\vec{\textrm{v}_f}}{2})\textrm{t} 
\end{align}



\begin{align}
\textrm{s} = \vec{\textrm{v}_i}\textrm{t}+\frac{1}{2}\vec{\textrm{a}}\textrm{t}^{2} \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  \ \ 
\vec{\textrm{v}_f} = \vec{\textrm{v}_i}^{2}+2 \vec{\textrm{a}}\textrm{s}
\end{align}

These are the four kinematic equations that are relevant to uniform motion, <font color = 'blue'> which we derive below. </font>

### Derivation of Equations of Motion

* $\vec{\textrm{v}_f}=\vec{\textrm{v}_i}+ \vec{\textrm{a}}\textrm{t}$

This equation defines the relation between velocity and acceleration. Suppose an object is moving in a constant acceleration $\textrm{a}$. From the definition of acceleration, we can write 

\begin{equation}
\vec{\textrm{a}}=\frac{\Delta \vec{\textrm{v}}}{\Delta \textrm{t}}
\end{equation}

where $\Delta \vec{\textrm{v}}$ is change of velocity over an _interval_ of time $\Delta \textrm{t}$, $\vec{\textrm{v}}_f - \vec{\textrm{v}}_i = \Delta \vec{\textrm{v}}$

If the object starts moving at $\textrm{t}=0$ and and initial velocity is $\vec{\textrm{v}_i}$, after some time $\textrm{t}$, the velocity has changed to be $\vec{v_f}$. The change of velocity is $\Delta \vec{\textrm{v}} = \vec{\textrm{v}_f} - \vec{\textrm{v}_i}$ and change of time is $\Delta \textrm{t} = \textrm{t}$. Then acceleration can be written as 

\begin{equation}
\vec{a}=\frac{\vec{\textrm{v}_f}- \vec{\textrm{v}_i}}{t} \\
\Rightarrow \vec{\textrm{v}_f} - \vec{\textrm{v}_i} = \vec{\textrm{a}}\textrm{t} \\
\Rightarrow \vec{\textrm{v}_f}= \vec{\textrm{v}_i}+ \vec{\textrm{a}} \textrm{t}
\end{equation}

* $\textrm{s} = (\frac{\vec{\textrm{v}_i}+\vec{\textrm{v}_f}}{2})\textrm{t}$

Average speed of an object, 
\begin{equation}
\textrm{average} \ \textrm{speed} = \frac{\textrm{distance} \ \textrm{covered}}{\textrm{time}} \\
\Rightarrow \frac{|\vec{\textrm{v}_i}|+|\vec{|\textrm{v}_f}|}{2}=\frac{\textrm{s}}{\textrm{t}} \\
\Rightarrow \textrm{s} = (\frac{|\vec{\textrm{v}_i}|+|\vec{\textrm{v}_f}|}{2})\textrm{t} 
\end{equation}

* $\textrm{s} = \vec{\textrm{v}_i}\textrm{t}+\frac{1}{2} \vec{\textrm{a}}\textrm{t}^{2}$

The displacement of an object 


Let an object travel with uniform acceleration $\textrm{a}$ with an initial velocity $\vec{\textrm{v}_i}$ $\textrm{ms}^{-1}$. At time $\textrm{t}$ $\textrm{sec}$, the object has covered a distance $\textrm{s}$ metres and a final velocity $\vec{\textrm{v}_f}$. If we write the definition for acceleration in mathematical terms:

\begin{equation}
\vec{\textrm{a}}=\frac{\vec{\textrm{v}_f}-\vec{\textrm{v}_i}}{t} \\
\Rightarrow {\vec{\textrm{v}_f}-\vec{\textrm{v}_i}}=\vec{\textrm{a}} \textrm{t} \\
\Rightarrow \vec{\textrm{v}_f}=\vec{\textrm{v}_i}+\vec{\textrm{a}}\textrm{t}
\end{equation}


The above equation represents the final velocity in terms initial velocity, acceleration and time. If we plot the graph of velocity as a function of time, it would appear as follows:

<img src="motion-graph.jpg" alt="Callysto" width=500 align = "left">

>*Note that this graph represents the relation between time verus velocity that we plotted earlier in uniform accelerated motion. As initial velocity $\textrm{v}_i$ is not zero which is acting as an intercept in the graph.* 

The graph of $\vec{\textrm{v}_f}=\vec{\textrm{v}_i}+\vec{\textrm{a}} \textrm{t}$ is a straight line where $\vec{\textrm{v}_i}$ as the $y$ intercept, $\vec{\textrm{a}}$ is the slope and $\textrm{t}$ is the independent variable. The area under the graph is the distance $\textrm{s}$ traveled by the object. To calculate the distance $\textrm{s}$, we split the graph into two parts: a triangle and a shaded rectangle. Let's calculate the area of triangle and shaded rectangle to obtain the displacement $\textrm{s}$,

**Area 1**
 
The area 1 represents the shaded rectangle. The area of a rectangle is the multiplication of length and height. In the graph, height represents the initial velocity $\vec{\textrm{v}_i}$ and length represents the time $\textrm{t}$. Let's calculate the area of the rectangle of the graph:

\begin{equation}
\textrm{Area}= \textrm{height} \times \textrm{length}= \vec{\textrm{v}_i}\textrm{t}
\end{equation}

**Area 2** 

The area 2 represents the area of the triangle. The area of a triangle relates to base and height. From the above graph, base is the time $t$ and the height is the difference between initial and final velocity. 

\begin{equation}
\textrm{Area}= \frac{1}{2} \times {\textrm{base}} \times {\textrm{height}} \\
\Rightarrow \textrm{Area}= \frac{1}{2} \times {\textrm{t}} \times {(\vec{\textrm{v}_f}-\vec{\textrm{v}_i)}} \\
\end{equation}

As we know, $(\vec{\textrm{v}_f}-\vec{\textrm{v}_i}) = \vec{\textrm{a}} \textrm{t}$. Replacing $(\vec{\textrm{v}_f}-\vec{\textrm{v}_i})$ in above equation:

\begin{equation}
\Rightarrow \textrm{Area}= \frac{1}{2} \times {t} \times \vec{a} {t} \\
\Rightarrow \textrm{Area} = \frac{1}{2} \times {\vec{a}t^{2}}
\end{equation}

The total displacement $s$ will be sum of area 1 and area 2.

\begin{equation}
\textrm{s}=\textrm{area 1} + \textrm{area 2} \\
\Rightarrow \textrm{s} = \vec{\textrm{v}_i}\textrm{t} + \frac{1}{2} {\vec{\textrm{a}}\textrm{t}^{2}}
\end{equation}

* $\vec{\textrm{v}_f} = \vec{\textrm{v}_i}^{2}+ 2 \vec{\textrm{ a}} \textrm{s}$

This equation is actually not anything new - it is just the combination of two of the equations we've already seen! We can get this equation by rearranging the the equations $\vec{\textrm{v}_f}=\vec{\textrm{v}_i}+ \vec{\textrm{a}}\textrm{t}$ and $\textrm{s} = \vec{\textrm{v}_f}\textrm{t}+\frac{1}{2}\vec{\textrm{a}}\textrm{t}^{2}$, as follows:

\begin{equation}
\vec{\textrm{v}_f}=\vec{\textrm{v}_i}+ \vec{\textrm{a}}\textrm{t} \\
\Rightarrow \vec{\textrm{v}_f}-\vec{\textrm{v}_i}=\vec{\textrm{a}}\textrm{t} \\
\Rightarrow \textrm{t}=\frac{\vec{\textrm{v}_f}-\vec{\textrm{v}_i}}{\vec{\textrm{a}}}
\end{equation}

Now, inserting the value of $\textrm{t}$ in displacement motion equation:

\begin{equation}
\textrm{s} = \vec{\textrm{v}_i} \textrm{t}+\frac{1}{2}\vec{\textrm{a}}\textrm{t}^{2} \\
\Rightarrow \textrm{s}= \vec{\textrm{v}_i}\frac{\vec{\textrm{v}_f}-\vec{\textrm{v}_i}}{\vec{\textrm{a}}}+ \frac{1}{2} (\frac{\vec{\textrm{v}_f}-\vec{\textrm{v}_i}}{t})^{2} \\
\Rightarrow \textrm{s} = \frac{1}{2} \frac{2\vec{\textrm{v}_i} \vec{\textrm{v}_f}-2\vec{\textrm{v}_i}^{2}+(\vec{\textrm{v}_f}-\vec{\textrm{v}_i})^{2}}{\vec{\textrm{a}}} \\
\Rightarrow \textrm{s} = \frac{1}{2} \frac{2\vec{\textrm{v}_i} \vec{\textrm{v}_f}-2\vec{\textrm{v}_i}^{2}+\vec{\textrm{v}_f}^{2}-2\vec{\textrm{v}_i} \vec{\textrm{v}_f}+\vec{\textrm{v}_i}^{2}}{\vec{\textrm{a}}} \\
\Rightarrow \textrm{s} =\frac{1}{2} \frac{\vec{\textrm{v}_f}^{2}-\vec{\textrm{v}}_i^{2}}{\vec{\textrm{a}}} \\
\Rightarrow \vec{\textrm{v}_f}^{2}-\vec{\textrm{v}_i}^{2}=2\vec{\textrm{a}}\textrm{s} \\
\Rightarrow \vec{\textrm{v}_f}^{2}=\vec{\textrm{v}_i}^{2}+2\vec{\textrm{a}}\textrm{s}
\end{equation}

#### Problem 1

A bus starting from accelerates from rest at $4  \textrm{ms}^{-2}$ until it reaches a final velocity of $40 \ \textrm{ms}^{-1}$. For how many seconds was the bus accelerating?

#### Solution

**Step 1:**

Given,

initial velocity, $\vec{\textrm{v}_i} = 0 \ \textrm{ms}^{-1}$ 

final velocity, $\vec{\textrm{v}_f} = 40 \ \textrm{ms}^{-1}$ 

acceleration, $\vec{\textrm{a}} = 4 \ \textrm{ms}^{-2}$

time, $\textrm{t} = ?$

**Step 2:**

In this problem, the value of $\vec{\textrm{v}_i}$, $\vec{\textrm{v}_f}$ and $\vec{\textrm{a}}$ are given and $\textrm{t}$ is required. Now, check the above $4$ motion equations, find the equation is related to$\vec{\textrm{v}_i}$, $\vec{\textrm{v}_f}$ and $\vec{\textrm{a}}$ and $\textrm{t}$. We find the equation and it is $\vec{\textrm{v}_f} = \vec{\textrm{v}_i}+\vec{\textrm{a}}\textrm{t}$.

**Step 3:** After checking all the units of the given variable $\textrm{v}_i$, $\textrm{v}_f$ and $\textrm{a}$, we find that they are in correct unit. Then fill them in selected equation:

\begin{equation}
\vec{\textrm{v}_f} = \vec{\textrm{v}_i}+\vec{\textrm{a}}\textrm{t} \\
\Rightarrow \vec{\textrm{a}}\textrm{t} = \vec{\textrm{v}_f} - \vec{\textrm{v}_i} \\
\Rightarrow \textrm{t} = \frac{\vec{\textrm{v}_f}-\vec{\textrm{v}_i}}{\vec{\textrm{a}}} \\
\Rightarrow \textrm{t} = \frac{40\textrm{ms}^{-1}-0\textrm{ms}^{-1}}{4\textrm{ms}^{-2}} \\
\Rightarrow \textrm{t} = 10\textrm{s}
\end{equation}

**Step 4:**

time = $10\textrm{s}$ $(Ans)$
